# Resolvendo problemas numéricos com python


A seguir, vamos mostrar como utilizar numpy e scipy para refazer as operações mostradas acima com sympy e visualizar os resultados usando a matplotlib.

Vamos pegar o exemplo acima das funções $x$, $x^2$ e $x^3$ que queremos plotar entre o intervalo $x \in [-2,2]$.

Importamos as bibliotecas
```python

import numpy as np # batizamos o numpy de np
import matplotlib.pyplot as plt # batizamos de plt
```

Podemos criar um conjunto de ```npts``` pontos associados à variável x no intervalo.
O código a seguir demonstra como fazer o mesmo plot anterior ilustrado com sympy de forma numérica.

`````python
import numpy as np
import matplotlib.pyplot as plt

npts = 101
x = np.linspace(-2,2,npts) # cria um conjunto de pontos igualmente espaçado

fig, ax = plt.subplots()

ax.axhline(0, color='black', linestyle='-', linewidth=0.5)
ax.axvline(0, color='black', linestyle='-', linewidth=0.5)

ax.plot(x, x, color='black', label=r'$x$')
ax.plot(x, x**2, color='red', label=r'$x^2$')
ax.plot(x, x**3, color='blue', label=r'$x^3$')

ax.set_xlabel(r'$x$') # label do eixo x
ax.set_ylabel(r'$f(x)$') # label do eixo y


leg = ax.legend()
`````

Podemos fazer várias coisas, por exemplo, somar as três funções com coeficientes arbitrários e ainda usar funções especiais (senos, cossenos, exponenciais) diretamente.

`````python
f1 = (2*x**3 - 10*x**2 + x) * np.sin(np.pi*x)
f2 = 4*np.exp(-0.5*x) * np.cos(4*np.pi*x)
fig2, ax2 = plt.subplots()

ax2.plot(x,f1, color='r', label='poly * sin')
ax2.plot(x,f2, color='g', label='damped oscillation')
ax2.set_ylabel(r'$f(x)$')
ax2.set_xlabel(r'$x$')
leg2 = ax2.legend()
`````

### Derivadas numéricas
Para funções de uma única variável $f(x)$ é  bacana usar o scipy.
A partir dos pontos $(x,y=f(x))$ guardados num ```numpy.array```, ele consegue criar uma função e derivá-la diretamente.

Disclaimer: existem maneiras mais sofisticadas de fazer derivadas numéricas. Comentaremos sobre isto ao longo do curso.


`````python
from scipy.interpolate import InterpolatedUnivariateSpline

# cria a função
fnum = InterpolatedUnivariateSpline(x, x**3-2*x, k=1)

# deriva a funcao
dfnum = fnum.derivative()
`````

`````python
fig3, ax3 = plt.subplots()
ax3.plot(x, fnum(x), color='blue',label=r'$f(x)$')
ax3.plot(x, dfnum(x), color='red', label=r"$f''(x)$ numerica ")
ax3.plot(x, 3*x**2 - 2, color='black', linestyle='--', label=r"$f''(x)$ analítica ")

leg3 = ax3.legend()
`````
